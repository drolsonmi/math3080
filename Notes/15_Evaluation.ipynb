## Evaluating Models

Our models don't do us much good if their predictions aren't very accurate. So, how do we evaluate the performance of our models?

The equations and methods of evaluation depend on the type of models.

### Evaluating Regression Models

The basic element of evaulation for regression models is the __residual__:

$$\hat{y}_i - y_i$$

This just tells us how far our prediction is from the true value. By itself, it isn't very useful. But in combination with the other residuals, it tells us a lot about the model.

#### Mean Absolute Error

$$MAE = \frac{\sum_i |\hat{y}_i - y_i|}{n}$$

The __mean absolute error__ (MAE) is a basic way to just take the average of the residuals. We do want to take the absolute value because when evaluating model performance with this equation, we really aren't interested in the direction of the residual (we are concerned, but not here). This is the simplest and most basic error measurement.

#### Sum of Squared Errors
The next step up is the __sum of squared errors__. In this case, we square the residual which does two things:
1. Minimizes small residuals so they don't add to the error
2. Amplifies (or __penalizes__) large residuals so they have a greater contribution to the error

$$SSE = \sum_i (\hat{y}_i - y_i)^2$$

If a model has only a couple of large residuals, there are only a couple of large "penalties" and the error remains relatively small. If there are a larger number of signficantly large residuals, then the penalties add up and the SSE grows significantly.

#### Mean Squared Error
The __Mean Squared Error__ (MSE) is one of the most commonly used error measures in data science. It is a continuation of the SSE but using it to find a mean.

$$MSE = \frac{\sum_i (\hat{y}_i - y_i)^2}{n}$$

#### Root Mean Squared Error
The MSE is extremely useful. However, the number isn't very interpretable because the difference is squared. To fix this problem, we take a square root. The result is the __root mean squared error__ (RMSE), and is another commonly used error measure.

$$RMSE = \sqrt{\frac{\sum_i (\hat{y}_i - y_i)^2}{n}}$$

#### Generalized Power Error
The MAE and the RMSE follow a generalized pattern. This pattern follows a generalized power error equation:

$$\left(\frac{\sum_i |\hat{y}_i - y_i|^k}{n}\right)^{1/k}$$

#### Root Mean Squared Log Error
Sometimes, the model follows an exponential pattern and is harder to evaluate. When this happens, the residuals might start scaling different orders of magnitude. So, it is sometimes useful to look at the model and the residuals on a logarithmic scale. The __root mean squared log error__ (RMSLE) helps evaluate an exponential model's performance:

$$RMSLE = \sqrt{\frac{\sum_i[\log(1 + \hat{y}_i) - \log(1 + y_i)]^2}{n}}$$

### Evaluating Categorical Models

__Confusion Matrix__

| *Total Population*   | Prediction Positive                   | Prediction Negative                    |
| -------------------: | :-----------------------------------: | :------------------------------------: |
| *Condition Positive* | True Positive (TP)                    | False Negative (FN)<br>(Type II error) |
| *Condition Negative* | False Positive (FP)<br>(Type I error) | True Negative (TN)                     |


Ultimately, either your model worked or it didn't. We can break this evaluation into multiple key metrics:
* Accuracy - Fraction of correct predictions
  $$ accuracy = \frac{\#~correct}{\#~of~predictions}$$
  * Good choice for balanced classes (equal numbers of different options)
  * Not a good choice for unbalanced classes
    * Predict 100 pictures to be dogs - If 99 are dogs, that's 99% accuracy!
* Recall
  * Ability of a model to find all relevant cases
  * Of all the items in a category, how many did I correctly predict?
$$recall = \frac{\#~true~positives}{\#~true~positives + \#~false~negatives}$$
* Precision
  * Ability to identify only the relevant cases
  * Of all the items I predicted in a category, how many were actually in that category?
$$precision = \frac{\#~true~positives}{\#~true~positives + \#~false~positives}$$
* F1-Score
  * Harmonic mean of precision and recall
$$F_1 = \frac{2\cdot precision\cdot recall}{precision + recall}$$

The method used depends on the situation.
* Situation determines if we fix the false positives or the false negatives

Let's look at a particular scenario. You have data from 1000 cancer patients that you run through a model to predict cancer diagnoses. The model gives you these results:

|                   | Positive Diagnosis | Negative Diagnosis |
| ----------------: | :----------------: | :----------------: |
| *Positive Cancer* | 214                | 18                 |
| *Negative Cancer* | 32                 | 736                |

$$Accuracy = \frac{214+736}{216+18+32+736} = \frac{950}{1000} = 0.95 = 95\%$$

$$Precision_+ = \frac{214}{214+32}=\frac{214}{246}=0.870=87.0\% \qquad Precision_- = \frac{736}{736+18} = \frac{736}{754} = 0.976 = 97.6\%$$

$$Recall_+ = \frac{214}{214+18} = \frac{214}{232} = 0.922 = 92.2\% \qquad Recall_- = \frac{736}{736+32} = \frac{736}{768} = 0.958 = 95.8\%$$

$$F_{1+} = \frac{2(0.870)(0.922)}{0.870+0.922}=0.895 \qquad F_{1-} = \frac{2(0.976)(0.958)}{0.976+0.958} = 0.967$$

```python
import numpy as np
from sklearn.metrics import classification_report

y_true = []
y_pred = []

for i in range(214):
    y_true.append(1)
    y_pred.append(1)
for i in range(18):
    y_true.append(1)
    y_pred.append(0) 
for i in range(32):
    y_true.append(0)
    y_pred.append(1) 
for i in range(736):
    y_true.append(0)
    y_pred.append(0) 

print(classification_report(y_true, y_pred))
```

Figures for the Ethics of Graphing lecture

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mpg = sns.load_dataset('mpg')
mpg['mpg'] = mpg['mpg'].astype(str)
mpg['horsepower'] = mpg['horsepower'].astype(str)
mpg.head()
```

```python
fig = plt.figure(figsize=(20, 20))

for country,col in zip(mpg['origin'].unique(),['red', 'green', 'blue']):
    plt.scatter(mpg[mpg['origin'] == country]['horsepower'], mpg[mpg['origin'] == country]['mpg'], color=col, label=country)

plt.xticks(rotation=90)
plt.savefig('../images/12_EthicsOfGraphing/bad_scatter.png')
```

```python
mpg['mpg'] = mpg['mpg'].astype(float)
mpg['horsepower'] = mpg['horsepower'].astype(float)

fig = plt.figure(figsize=(8, 8))

for country,col in zip(mpg['origin'].unique(),['red', 'green', 'blue']):
    plt.scatter(mpg[mpg['origin'] == country]['horsepower'], mpg[mpg['origin'] == country]['mpg'], color=col, label=country)

plt.xticks(rotation=90)
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon')
plt.title('MPG vs Horsepower by Country of Origin')
plt.legend()
plt.savefig('../images/12_EthicsOfGraphing/good_scatter.png')
```
